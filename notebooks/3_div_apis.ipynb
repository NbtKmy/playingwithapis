# Verschiedene Arten von APIs – Ein Überblick

In diesem Abschnitt stellen wir einige interessante **APIs** vor, die für Forschung, Datenanalyse und kreative Projekte genutzt werden können. Es gibt sowohl **öffentlich zugängliche APIs** als auch solche, die **von Universitäten oder Forschungseinrichtungen bereitgestellt** werden.

---

## Öffentlich verfügbare APIs

Diese APIs können in der Regel frei oder mit einfacher Registrierung verwendet werden. Sie eignen sich hervorragend für den Einstieg und Experimente.

- **[Free Public APIs – Showcase](https://www.freepublicapis.com/tags/showcase)**  
  Eine Sammlung von offenen APIs zu verschiedenen Themen – von Kunst und Wissenschaft bis hin zu Wetterdaten und Übersetzungsdiensten.

- **Bluesky API**  
  Bietet Zugang zur dezentralen Social-Media-Plattform *Bluesky*. Mit dieser API kann man Posts abrufen, Nutzer*innenprofile analysieren oder eigene Anwendungen rund um das AT-Protokoll entwickeln.

- **YouTube Data API**  
  Ermöglicht den Zugriff auf Metadaten zu Videos, Kanälen und Playlists auf YouTube. Ideal für Studien zu Mediennutzung, Algorithmusanalyse oder für eigene Visualisierungsprojekte.

- **Europeana API**  
  Stellt Millionen von digitalen Objekten aus europäischen Museen, Archiven und Bibliotheken bereit – darunter Bilder, Texte, Audio- und Videodateien. Ein wertvolles Werkzeug für Digital Humanities und Kulturforschung.


- **[OpenSitations API](https://api.opencitations.net/index/v2)**

## APIs von kommerziellen Anbietern

Es sind auch APIs von kommerziellen Anbietern. Das heisst, dass der Zugriff auf eine API-Schnittstelle kostenpflichtig ist. Ein Paar Beispiele hier sind:

- **X API v2 Full Archive Search** - Full Archive Search ist nur für Pro- bzw. Enterprise-User freigegeben
- **OpenAI API** - Zugriff auf LLM oder sonstige KI-Modelle wird nach Gebrauch gemessen und demnach die Gebühren erhoben
- **[List of news media apis](https://en.wikipedia.org/wiki/List_of_news_media_APIs)**
- **[Open Router API](https://openrouter.ai/)**
- **[Poe API](https://poe.com/blog/introducing-the-poe-api)**

- **LexisNexis API**  
  Ein professioneller Zugang zu internationalen Nachrichten-, Rechts- und Wirtschaftsdaten. Besonders nützlich für Forschungsprojekte in Medienwissenschaften, Jura und Wirtschaft. UB Zürich bietet den Zugang für die UZH-Angehörigen an.

## Universitäre und institutionelle APIs

Diese APIs werden oft von Forschungseinrichtungen oder Universitäten bereitgestellt und bieten spezialisierte Datenbestände für wissenschaftliche Analysen.


- **[The Swissdox@LiRI API](https://liri.linguistik.uzh.ch/wiki/langtech/swissdox/api)**  
  Entwickelt an der Universität Zürich, ermöglicht diese API den Zugriff auf umfangreiche Schweizer Pressedaten. Forschende können damit Textanalysen, linguistische Studien oder Medienvergleiche durchführen.

---

## Fazit

APIs eröffnen vielfältige Möglichkeiten, **digitale Informationen strukturiert abzurufen und zu analysieren**.  
Ob für kleine Experimente, Forschungsprojekte oder innovative Anwendungen – der kreative Umgang mit Daten steht im Mittelpunkt.
